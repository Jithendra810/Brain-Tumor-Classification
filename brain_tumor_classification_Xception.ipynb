```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import Xception
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, Callback
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
# Define dataset paths
data_dir = "archive"
train_dir = os.path.join(data_dir, "training")
test_dir = os.path.join(data_dir, "testing")
```

```python
# Image size and batch size
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
#  Advanced Data Augmentation (Prevents Overfitting)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,  
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3, 
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],  
    fill_mode='nearest',
    validation_split=0.2  # 20% for validation
)
```

```python
# Load training and validation data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Load test data
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
```

```python
#  Fine-Tuned Xception Model (Higher Accuracy)
def build_model():
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    
    # Unfreeze top layers for fine-tuning
    for layer in base_model.layers[:-30]:  
        layer.trainable = False
    for layer in base_model.layers[-30:]:  
        layer.trainable = True
    
    inputs = tf.keras.Input(shape=(224, 224, 3))
    x = base_model(inputs, training=True)
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)  # Normalize activations for stability
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)  # Increased dropout to reduce overfitting
    outputs = Dense(4, activation='softmax')(x)  

    model = Model(inputs, outputs)
    return model

# Create model
model = build_model()
```

```python

#  Use AdamW Optimizer with Lower Learning Rate
model.compile(optimizer=AdamW(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
print("Model compiled successfully.")
```

```python
#  Callbacks to Improve Accuracy
checkpoint = ModelCheckpoint("best_Xception.keras", save_best_only=True, monitor='val_accuracy', mode='max')
early_stopping = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)


class StopWhenAccuracyReached(Callback):
    def __init__(self, target_accuracy=0.97):
        super(StopWhenAccuracyReached, self).__init__()
        self.target_accuracy = target_accuracy

    def on_epoch_end(self, epoch, logs=None):
        if logs.get("accuracy") >= self.target_accuracy:
            print(f"\nüöÄ Accuracy reached {self.target_accuracy * 100:.2f}%, stopping training at epoch {epoch + 1}!")
            self.model.stop_training = True  # Stop training immediately

            # Stop further echoing (printing logs)
            import sys
            sys.stdout = open(os.devnull, 'w')  # Redirect stdout to null (disable printing)
```

```python
# üî• Train the Model (Increased Epochs)
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=100,  # Increased to 100
    callbacks=[checkpoint, early_stopping, reduce_lr,stop_accuracy]
)

print("Training complete.")
```

```python
# üî• Evaluate the Model on Test Data
model = load_model("best_Xception.keras")
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy:.4f}, Test Loss: {loss:.4f}")
```
